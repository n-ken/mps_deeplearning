## sigmoid関数

$$f(x) = \frac{1}{1 + e^{-x}}$$

採用した理由:

- ステップ関数の近似となる
- 微分係数の表現が簡単になる

sigmoid関数の性質: 

- 出力が常に正である。
- 微分係数の最大値が0.25である。
- $x$の絶対値が大きくなるほど、勾配が0に近づく。

学習への影響: 

- sigmoid関数の出力が、常に正となる問題点

    - 勾配$\frac{\partial E}{\partial w_{ji}^{l}} = \delta_{j}^{l} y_{ji}^{(l-1)}$のうち、$y_{ji}^{(l-1)}$が必ず正となる。
    
        - $\delta_{j}^{l}$は共通で使われる。かつ$y_{ji}^{(l-1)}$は必ず正である。すると、勾配の符号が必ず同じになる。
        - すなわち、勾配の方向が制限される。(勾配の要素の符号が必ず同じになる。)
        
- 微分係数の最大値が0.25である問題点

    - $\delta$の更新式(BackPropagation)の過程で微分係数が使用される。
    - すなわち、$\delta$は前の層に遡るにつれて小さくなっていく。
